```python
class Atm: 
# constructor 
    def __init__(self):#Non parameterize constructor(not required parameter) 
        print(id(self))
        self.pin='123' 
        self.balance=10000
        # self.menu() 
    def menu(self): 
        user_input=input('''
         Hello, 
         How can I help you! 
         Please Choose Any one option.
         1.Press 1 to Balance Inquiry
         2.Press 2 to Case withdraw
         3.Press 3 to Pin Change
         4.Press 4 to Create Pin
         5.Any things else to exist''')
        if user_input=='1': 
            self.Balance_inquiry()
        elif user_input=='2':
            self.case_withdraw()
        elif user_input=='3': 
            self.Change_pin()
        elif user_input=='4': 
            self.create_pin() 
        else:
            exit()
    def create_pin(self): 
        user_pin=input("Enter Your Pin:")
        self.pin=user_pin
        user_balance=input("Enter Your Balance:")
        self.balance=user_balance
        print("pin created successfully")
        self.menu()
    def Change_pin(self): 
        old_pin=input("Enter your old pin:")
        
        if old_pin==self.pin: 
            new_pin=input("Enter new pin:")
            self.pin=new_pin
            print("New pin created successful")
            self.menu()
        else:
            print("Incorrect pin Please try again....")
            self.menu()
    def Balance_inquiry(self): 
        user_pin=input("Enter Your pin:")
        if user_pin==self.pin: 
            print("Your Balance is:",self.balance)
            self.menu()
        else: 
            print("Incorrect pin Please try again....")
            self.menu() 
    def case_withdraw(self): 
        user_pin=input("Enter Your pin:")
        if user_pin==self.pin: 
            req_bal=int(input("Enter Amount:"))
            if req_bal<=self.balance: 
                print("Collect Your Cash:")
                self.balance=self.balance-req_bal
                print("Remaing Balance is:",self.balance)
                self.menu()
            else: 
                print("Insufficient balance.....")
                self.menu()
        else:
            print("Incorrect pin please Try again.....")
            self.menu()
    
```

```python
class fraction: 
    def __init__(self,x,y): 
        self.num=x
        self.den=y
    # def __str__(self): 
    #     return '{}/{}'.format(self.num,self.den)
    def __add__(self,other): 
        new_num=self.num*other.den+other.num*self.den
        new_dem=self.den*other.den
        return '{}/{}'.format(new_num,new_dem)
    def __sub__(self,other): 
        new_num=self.num*other.den-other.num*self.den
        new_dem=self.den*other.den
        return '{}/{}'.format(new_num,new_dem) 
    def __mul__(self,other): 
        new_num=self.num*other.num
        new_dem=self.den*other.den
        return '{}/{}'.format(new_num,new_dem)
    def __truediv__(self,other): 
        new_num=self.num*other.den
        new_dem=self.den*other.num
        return '{}/{}'.format(new_num,new_dem)
    def convert_to_decimal(self):
        return self.num/self.den
        
    
        
```

```python
# writer a program to check whether a number is prime or not. 
n=int(input('enter an Integer no.'))
if n<=1: 
    print(n,'is not a prime no.')
else:
    for i in range(2,n): 
        if n%i==0:
            print(n,'is not a prime no.')
            break
    else:
        print(n,'is prime no.')
```

```python
# find and print the largest and smallest no. in a list [8,2,15,1,9] without using max() and min().  
ls=[8,2,15,1,9]
largest=ls[0] 
smallest=ls[0] 
for i in ls:
    if i >largest: 
        largest=i
    elif i<smallest:
        smallest=i
print('largest no.',largest)
print('smallest no.',smallest)
```
